```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as  sns
```

```python

dataset=pd.read_csv('train (1).csv')
dataset.head()
```

```python
features_with_na=[a for a in dataset.columns if dataset[a].isnull().sum()!=0]
```

```python
dataset['LotFrontage'].isnull().sum()
```

```python
features_with_na
```

```python
for feature in features_with_na:
    print(feature,dataset[feature].isnull().mean())
```

```python
for feature in features_with_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.title(feature)
    plt.show()
    
```

```python
for feature in year_feature:
    if feature!= 'YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique()) < 25 and feature not in year_feature+['Id']]
len(discrete_feature)
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].mean().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
continuous_feature=[feature for  feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
continuous_feature
```

```python
len(continuous_feature)
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else :
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```
