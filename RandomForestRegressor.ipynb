```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
diabetes=datasets.load_diabetes(as_frame=True)
diabetes.keys()
```

```python
print(diabetes.DESCR)
```

```python
x=diabetes.data
y=diabetes.target
y
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
regressor=RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
```

```python
plt.figure(figsize=(16,4))
plt.plot(y_pred,label="y_pred")
plt.plot(y_test,label="y_test")
plt.xlabel("x_test",fontsize=14)
plt.ylabel("value of y(pred,test)",fontsize=14)
plt.title("comparing predicted values and true values")
plt.legend(title="parameter ehere:")
plt.show()
```

```python
np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```

```python
iris=datasets.load_iris()
iris.target_names
```

```python
x=iris.data
y=iris.target
```

```python
clf=RandomForestClassifier(n_estimators=100,random_state=4)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
metrics.accuracy_score(y_test,y_pred)
```

```python
mat=metrics.confusion_matrix(y_test,y_pred)
```
